```python
### Let's crate a Bar Graph
fig,ax = plt.subplots(figsize = (5,5))
x = ["Oil", "Shampoo", "Soap"]
y = [2211, 775, 555]
ax.bar(x,y)
```

```python
### Let's crate a Bar Graph..........add more information.....
fig,ax = plt.subplots(figsize = (5,5))
x = ["Oil", "Shampoo", "Soap"]
y = [1284, 775, 555]
ax.bar(x,y)
ax.set_xlabel("Products")
ax.set_ylabel("Units Sold")
ax.set_title("Units sold for each product")
ax.grid(True)                                ### Shows Grid
```

```python
### Let's crate a Bar Graph
fig,ax = plt.subplots(figsize = (5,5))
x = ["Oil", "Shampoo", "Soap"]
y = [1284, 775, 555]
colour_bar = ["Blue","Green","Yellow"]

ax.bar(x,y, colour = colour_bar)
ax.set_xlabel("Products")
ax.set_ylabel("Units Sold")
ax.set_title("Units sold for each product")
ax.grid(True)                                ### Shows Grid
```

```python
### Let's crate a Bar Graph...........Change colours
fig,ax = plt.subplots(figsize = (5,5))
x = ["Oil", "Shampoo", "Soap"]
y = [1284, 775, 555]
color_bar = ["Light Blue","Green","Yellow"]

ax.bar(x,y, color = color_bar)
ax.set_xlabel("Products")
ax.set_ylabel("Units Sold")
ax.set_title("Units sold for each product")
ax.grid(True)                                ### Shows Grid
```
