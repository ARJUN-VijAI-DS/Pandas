<font size = 5.5> Scatter Plot  ---- Relationship.... Linear, Inverse or No Effect (No relation)....

```python
### Scatter Plot
x = np.linspace(1,20)
y = [i**2 for i in x]
y1 = [i**3 for i in x]

fig,ax = plt.subplots(figsize = (5,5))
ax.scatter(x,y, alpha = 0.7, color = "green" , marker= "*")
ax.scatter(x,y1, alpha = 0.8, color = "blue" , marker= "o")
ax.legend(["**2", "**3"])
ax.grid(True)
```

```python
### Line Plot/Graph
x = np.linspace(1,20)
y = [i**2 for i in x]
y1 = [i**3 for i in x]

fig,ax = plt.subplots(figsize = (5,5))
ax.plot(x,y, alpha = 0.9, color = "green" , marker= "*")     #### alpha is controlling the color darkness. 0 = no color
ax.plot(x,y1, alpha = 1, color = "blue" , marker= "o")        ### alpha  = 1 means maximum darkness..... 
ax.legend(["**2", "**3"])
ax.grid(True)
```

```python
### Line Plot/Graph
x = np.linspace(1,20)
y = [i**2 for i in x]
y1 = [i**3 for i in x]
y2 = [i**4 for i in x]
fig,ax = plt.subplots(figsize = (5,5))
ax.plot(x,y, alpha = 0.9, color = "green")       #### alpha is controlling the color darkness. 0 = no color
ax.plot(x,y1, alpha = 1, color = "blue" )        ### alpha  = 1 means maximum darkness..... 
ax.plot(x,y2, alpha = .8, color = "red" )
ax.legend(["**2", "**3", "**4"])
ax.grid(True)
```

```python
### Line Plot/Graph
x = np.linspace(1,5)
y = [i**2 for i in x]
y1 = [i**3 for i in x]
y2 = [i**4 for i in x]
fig,ax = plt.subplots(figsize = (5,5))
ax.plot(x,y, alpha = 0.9, color = "green")       #### alpha is controlling the color darkness. 0 = no color
ax.plot(x,y1, alpha = 1, color = "blue" )        ### alpha  = 1 means maximum darkness..... 
ax.plot(x,y2, alpha = .8, color = "red" )
ax.set_xlabel("Marketing Expenditure")
ax.set_ylabel("Sales")
ax.legend(["**2", "**3", "**4"])
ax.grid(True)
```
